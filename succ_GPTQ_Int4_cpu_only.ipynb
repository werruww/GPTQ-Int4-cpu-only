```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# اسم النموذج من Hugging Face
model_name = "openai-community/gpt2"  # أو أي نموذج آخر تدعمه Hugging Face

# تحميل النموذج بدقة 4 بت
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    load_in_4bit=True,  # تفعيل تحميل النموذج بدقة 4 بت
    device_map="cpu",   # استخدام المعالج فقط
    torch_dtype=torch.float32
)

# تحميل Tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# إدخال النص
input_text = "مرحبًا، كيف يمكنني مساعدتك اليوم؟"

# تحويل النص إلى tokens
inputs = tokenizer(input_text, return_tensors="pt")

# توليد النص
outputs = model.generate(**inputs, max_length=50)

# تحويل الناتج إلى نص
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```

```python
from transformers import pipeline
import torch

# تعريف الرسائل
messages = [
    {"role": "user", "content": "Who are you?"},
]

# إنشاء pipeline مع تحديد torch_dtype و device_map
pipe = pipeline(
    "text-generation",
    model="unsloth/Llama-3.2-11B-Vision-Instruct-bnb-4bit",
    trust_remote_code=True,
    torch_dtype=torch.float32,  # استخدام دقة 16 بت لتقليل استخدام الذاكرة
    device_map="auto"           # تحديد الجهاز تلقائيًا (CPU أو GPU)
)

# تشغيل النموذج
outputs = pipe(messages)

# عرض الناتج
print(outputs)
```

```python
from transformers import pipeline
import torch

# تعريف الرسائل
messages = [
    {"role": "user", "content": "Who invented the light bulb?"},
]

# إنشاء pipeline مع تحديد torch_dtype و device_map
pipe = pipeline(
    "text-generation",
    model="Qwen/Qwen2.5-Coder-7B-Instruct-GPTQ-Int4",
    trust_remote_code=True,
    torch_dtype=torch.float32,  # استخدام دقة 16 بت لتقليل استخدام الذاكرة
    device_map="auto"           # تحديد الجهاز تلقائيًا (CPU أو GPU)
)

# تشغيل النموذج
outputs = pipe(messages, max_new_tokens=5)

# استخراج النص المولد فقط
generated_text = outputs[0]['generated_text']

# عرض الناتج
print(generated_text)
```
